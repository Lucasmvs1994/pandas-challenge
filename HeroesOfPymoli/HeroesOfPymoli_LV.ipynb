```python
# Age Demographics Analysis 


print(data_df["Age"].max())
print(data_df["Age"].min())
```

```python

bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000 ]

# Creating  names for the bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]
```

```python
pd.cut(data_df["Age"], bins, labels=group_names).head()
```

```python
data_df["Age Group"] = pd.cut(data_df["Age"], bins, labels=group_names)
data_df.head()
```

```python
age_group_df = data_df.groupby("Age Group")

print(age_group_df['SN'].nunique())
```
